```python
import sys

sys.path.append('/content/DP_Lab')
import pda
```

```python
#write a path to the pda file txt
pda_file = '/content/DP_Lab/sample_data/sample_pda_2.txt'
#open the file, 'name' will be the title of the plot
mydata = pda.PDA(pda_file, name='PC-Tz 100 mM')
```

```python
#plot chromatogram at different wavelength
mydata.plot_at_wl(280, 500)
```

```python
#plot pda intensity at multiple time points
mydata.plot_at_rt(13, 13.5, colors=['blue', 'red'])
```

```python
#detect peak so you don't have to guess the time points of the peaks
mydata.detect_peak(wl=280) #wl: choose which wavelength from which you want to detect the peaks
```

```python
#since you have registered peaks, not you can just call which peak you want to plot
mydata.plot_peaks(1, 2) 
```

```python
#let's see closer to the spectra around 500 by setting the x and y axis limit
mydata.plot_peaks(1, 2, xlim=(450, 600), ylim=(-1e2, 5e3)) 
```

```python
#you can also change the colors
mydata.plot_peaks(1, 2, xlim=(450, 600), ylim=(-1e2, 5e3), colors=['blue', 'red']) 
```
